# Visualizing the Gender Gap in College Degrees
The Department of Education Statistics releases a data set annually containing the percentage of bachelor's degrees granted to women from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category. <br>

Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. You can download the dataset Olson compiled [here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv). <br>

In this project, we will try to explore the gender gap in college degrees using effective data visualization. 

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
women_degrees.head()
```

First, we will visualize the percentages of Bachelor's degrees in STEM fields:

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

Next, we will visualize the percentages of Bachelor's degrees in STEM, liberal arts, and other fields.

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
fig = plt.figure(figsize=(16, 20))
fig.suptitle("Visualizing The Gender Gap In College Degrees",fontsize=20)

#generate the STEM majors' column
for sp in range(1,17,3):
    ax = fig.add_subplot(6,3,sp)
    index=int((sp-1)/3)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if index == 0:
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')

#generate the liberal arts' majors column
for sp in range(2,15,3):
    ax = fig.add_subplot(6,3,sp)
    index=int((sp-2)/3)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if index == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
        
#generate the other majors column
for sp in range(3,19,3):
    ax = fig.add_subplot(6,3,sp)
    index=int((sp-3)/3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if index == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        
fig.tight_layout()
fig.subplots_adjust(top=0.94)
plt.show()
```

# Hiding x-axis labels
With seventeen lie charts in one diagram, we notice that the non-data elements are cluttering the field of view. We notice that the x-axis labels are overlapping with the titles of the following charts. <br>
We will remove the x-axis label for all the diagrams and keep them for the bottom diagrams of each column.

# Adding a horizontal line across the plots
Because we removed most of the y-axis labels, we indeed reduced the visual clutter, but made it hard to discern which degrees have close to a 50-50 gender breakdown.
We will generate a horizontal line across each plot where the y-axis label 50 would have been.

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
fig = plt.figure(figsize=(16, 20))
fig.suptitle("Visualizing The Gender Gap In College Degrees",fontsize=20)

#generate the STEM majors' column
for sp in range(1,17,3):
    ax = fig.add_subplot(6,3,sp)
    index=int((sp-1)/3)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False,labelbottom=False)
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255, 171/255, 171/255), alpha=0.3) #generate a horizontal line at 50%
    if index == 0:
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
        ax.tick_params(labelbottom=True)

#generate the liberal arts' majors column
for sp in range(2,15,3):
    ax = fig.add_subplot(6,3,sp)
    index=int((sp-2)/3)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False,labelbottom=False)
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255, 171/255, 171/255), alpha=0.3) #generate a horizontal line at 50%
    if index == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
    if index==4:
        ax.tick_params(labelbottom=True)
        
#generate the other majors column
for sp in range(3,19,3):
    ax = fig.add_subplot(6,3,sp)
    index=int((sp-3)/3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False,labelbottom=False)
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255, 171/255, 171/255), alpha=0.3) #generate a horizontal line at 50%
    if index == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        ax.tick_params(labelbottom=True)
        
fig.tight_layout()
fig.subplots_adjust(top=0.94)
plt.show()
```
